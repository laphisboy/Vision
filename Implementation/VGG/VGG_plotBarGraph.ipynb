```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
label_dataset = ['CIFAR10', 'Imagewoof', 'Imagewoof w. Altered VGG']
```

```
accuracy1 = [0.8477, 0.3723, 0.4625] # Pretraining
accuracy2 = [0.9166, 0.1034, 0.4953] # Aftertraining
```

```
x_dataset = np.arange(len(label_dataset))

width = 0.35

fig1, ax1 = plt.subplots()
plt.rcParams["figure.figsize"] = (10, 7)
COLOR = 'white'
plt.rcParams['text.color'] = 'black'
plt.rcParams['axes.titlecolor'] = COLOR
plt.rcParams['axes.labelcolor'] = COLOR
plt.rcParams['xtick.color'] = COLOR
plt.rcParams['ytick.color'] = COLOR

rects1 = ax1.bar(x_dataset - width/2, accuracy1, width/4, label='pre')
rects2 = ax1.bar(x_dataset + width/2, accuracy2, width/4, label='74 epoch')

ax1.set_ylabel('Accuracy')
ax1.set_title('Graph 1: Comparison of Accuracy for Different Datasets', pad=20)
ax1.set_xticks(x_dataset)
ax1.set_xticklabels(label_dataset)
ax1.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax1.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig1.tight_layout()
plt.show()
```
